## Popularity Based Recommand System

```python
book_rating = rating.merge(book,on = 'ISBN')
```

```python
num_rating = book_rating.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
avg_rating = book_rating.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating' : 'avg-rating'},inplace =True)
avg_rating
```

```python
rating_df = num_rating.merge(avg_rating,on = 'Book-Title')
rating_df
```

```python
top_popular = rating_df[rating_df['Book-Rating'] >=250].sort_values('avg-rating', ascending = False).head(40)
top_popular
```

```python
popular_books = top_popular.merge(book,on = 'Book-Title').drop_duplicates("Book-Title")[['Book-Title','Book-Author','Image-URL-M','Book-Rating', 'avg-rating' ]]
```

```python
popular_books
```

## Collaborative Filtering Based Recommand System

```python
x = book_rating.groupby('User-ID').count()['Book-Rating'] > 200
good_user = x[x].index
good_user
```

```python
filter_rating = book_rating[book_rating['User-ID'].isin(good_user)]
```

```python
y = filter_rating.groupby('Book-Title').count()['Book-Rating'] >= 40
famous_book = y[y].index
```

```python
famous_book
```

```python
final_df = filter_rating[filter_rating['Book-Title'].isin(famous_book)]
```

```python
pt = final_df.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt.fillna(0, inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar_score = cosine_similarity(pt)
similar_score
```

```python
def recommand(book_name):
    index = np.where(pt.index== book_name)[0][0]
    similar_items = sorted(list(enumerate(similar_score[index])),key = lambda x:x[1],reverse = True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = book[book['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
```

```python
recommand('1984')
```
